```python
import json
import matplotlib.pyplot as plt
import statistics

DELTAS: list = [1, 3, 5, 7]


data = json.load(open("/local/cse592_chess/data.json"))

d1_1200 = data['1200']['1']['data']

sf_avgs: list = []
lc_avgs: list = []

for delta in DELTAS:
    sf_data, lc_data = zip(*data['1200'][str(delta)]['data'])

    sf_data = list(sf_data)
    lc_data = list(lc_data)

    sf_avgs.append(statistics.mean(sf_data))
    lc_avgs.append(statistics.mean(lc_data))
```

```python
plt.plot(DELTAS, sf_avgs, label="Stockfish")
plt.plot(DELTAS, lc_avgs, label="Lc0")

plt.title("Evaluation Accuracy (ELO = 1200)")
plt.xlabel("Delta")
plt.ylabel("Evaluation Error (centipawns)")

plt.grid(True)
plt.legend()
plt.show()
```
